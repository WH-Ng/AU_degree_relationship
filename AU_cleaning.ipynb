#### Comments
- A lot of program electives - 'major' or 'year x' or 'Business'. These are good to have, but may not be important with what i want to find out.

#### Questions:
- What is the difference between 'Core courses', 'Core selective courses', 'Common core', 'Program core', 'Program core selectives'
- How many degrees overlap with core courses?

#### Answers
- Common core are courses that compulsory for ALL undergraduate degrees. Lets leave this one out.
- Two key streams: Program Core and Core Selectives (need to do mapping)

```python
acc_section = accounting2[accounting2['section'] == 'Core courses'].copy()
```

```python
acc_section.stream.unique()
```

```python
acc_section
```

```python
def map_stream(s):
    if pd.isna(s):
        return 'Program Core'
    s_lower = s.strip().lower()
    
    # Program Core variants
    if "program core" in s_lower or s_lower.startswith("program core") or s_lower.startswith("Program core selectives") or s_lower.startswith("'Program core - selective'"):
        return "Program Core"
        
    # Core selectives / core selective
    if "core courses" in s_lower or s_lower.startswith("Core selective courses"):
        return "Core Selectives"
        
    # Common core — treat separately (we'll exclude later if you want)
    if "common core" in s_lower or s_lower.startswith("common core"):
        return "Common Core"
        
    # fallback: return original trimmed
    return s.strip()
```

```python
acc_section['stream_clean'] = acc_section['stream'].apply(map_stream)
```

```python
acc_section
```

```python
acc_core_filtered = acc_section[acc_section['stream_clean'].isin(['Program Core','Core Selectives'])].copy()
```

```python
acc_core_filtered.stream_clean.unique()
```

```python
acc_degrees = acc_core_filtered[['degree','course_code','course_name', 'stream_clean']].rename(columns={'degree':'degree_name'})
```

```python
acc_degrees.course_code.nunique()
```

```python
acc_degrees.to_csv("degree_courses_clean.csv", index=False)
```

```python
course_freq = acc_degrees.groupby(['course_code','course_name'])['degree_name'].nunique().reset_index()
course_freq = course_freq.rename(columns={'degree_name':'degree_count'}).sort_values('degree_count', ascending=False)
```

```python
course_freq.to_csv("course_frequency.csv", index=False)
```

```python
course_freq
```

```python
#%pip install pandas numpy networkx

import pandas as pd
import numpy as np
from itertools import combinations
import networkx as nx
import json

degrees_acc = sorted(acc_degrees['degree_name'].unique())
degree_to_courses = {d: set(acc_degrees[acc_degrees['degree_name']==d]['course_code']) for d in degrees_acc}

rows = []
for a,b in combinations(degrees_acc,2):
    shared = degree_to_courses[a].intersection(degree_to_courses[b])
    shared_count = len(shared)
    if shared_count >= 1:
        rows.append({'degree_a':a, 'degree_b':b, 'shared_count':shared_count, 'shared_courses': ", ".join(sorted(shared))})

acc_overlap = pd.DataFrame(rows).sort_values('shared_count', ascending=False)
```

```python
acc_overlap.to_csv("degree_overlap_pairs.csv", index=False)
```

```python
acc_overlap
```

```python
import numpy as np
idx = {d:i for i,d in enumerate(degrees_acc)}
mat = np.zeros((len(degrees_acc), len(degrees_acc)), dtype=int)
for _,r in acc_overlap.iterrows():
    i,j = idx[r['degree_a']], idx[r['degree_b']]
    mat[i,j] = r['shared_count']
    mat[j,i] = r['shared_count']

acc_df_mat = pd.DataFrame(mat, index=degrees_acc, columns=degrees_acc)
```

```python
acc_df_mat.to_csv("degree_overlap_matrix.csv")
```

```python
acc_df_mat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(acc_df_mat, annot=True, cmap='coolwarm', square=True, linewidths=0.5, cbar_kws={"shrink": 0.5})
plt.title('Heatmap of Degree Overlap Matrix')
plt.show()
```

```python
G = nx.Graph()
for d in degrees_acc:
    G.add_node(d)
for _,r in acc_overlap.iterrows():
    G.add_edge(r['degree_a'], r['degree_b'], weight=int(r['shared_count']))

# Export edges + nodes
graph_out = {
    'nodes': [{'id': n} for n in G.nodes()],
    'edges': [{'source': u, 'target': v, 'weight': int(d['weight'])} for u,v,d in G.edges(data=True)]
}
with open("degree_graph.json","w") as f:
    json.dump(graph_out, f, indent=2)
print("Saved degree_graph.json")
```
